```python
import pandas as pd

# Load ZHVI data
data = pd.read_csv("zillow_HVI_data.csv")
```

```python
# fill missing values with median
data.fillna(data.median(), inplace=True)
```

```python
# Transpose the DataFrame
data_transposed = data.transpose()
```

```python
#reset the index
data_transposed.reset_index(inplace=True)
```

```python
#rename col names
data_transposed.columns = data_transposed.iloc[0]
data_transposed = data_transposed[1:]
```

```python
#save preprocessed data
data_transposed.to_csv("preprocessed_data.csv", index=False)
```
